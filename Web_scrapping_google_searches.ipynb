```python
import requests
from bs4 import BeautifulSoup
import urllib.parse

def get_google_search_links(query, num_results=10):
    query = urllib.parse.quote_plus(query)
    url = f"https://www.google.com/search?q={query}&num={num_results}"

    #headers = {
        #"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.6422.176 Safari/537.3"}

    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        links = []
        for item in soup.find_all('a'):
            href = item.get('href')
            if href and '/url?q=' in href:
                link = href.split('/url?q=')[1].split('&sa=U')[0]
                links.append(urllib.parse.unquote(link))
        return links
    else:
        return []

# Example usage:
query = "sattva infrastructure india private limited"
links = get_google_search_links(query)
for link in links:
    print(link)
```

```python
import requests
from bs4 import BeautifulSoup

def get_text_from_links(links):

    link_texts = {}

    for link in links:
        try:
            response = requests.get(link)
            if response.status_code == 200:
                soup = BeautifulSoup(response.content, 'html.parser')
                text = soup.get_text(separator='\n', strip=True)
                link_texts[link] = text
            else:
                link_texts[link] = f"Failed to retrieve: Status code {response.status_code}"
        except requests.RequestException as e:
            link_texts[link] = f"Error: {e}"

    return link_texts

# Example usage:
# links = [
#     "https://www.example.com",
#     "https://www.wikipedia.org",
#     # Add more links as needed
# ]

link_texts = get_text_from_links(links)

# Print the extracted text for each link
for link, text in link_texts.items():
    print(f"Text from {link}:\n{text}\n{'-'*80}\n")
```

```python
import requests
from bs4 import BeautifulSoup

def get_heading_and_text_from_links(links):
    headers = {
         "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.6422.176 Safari/537.3"
     }

    link_contents = {}

    for link in links:
        try:
            session = requests.Session()
            response = session.get(link, headers=headers, verify=False)
            #response = requests.get(link, headers=headers, verify=False)
            if response.status_code == 200:
                soup = BeautifulSoup(response.content, 'html.parser')

                # Extract the title
                title = soup.title.string if soup.title else 'No title found'

                # Extract headings and paragraphs
                headings = soup.find_all(['h1', 'h2', 'h3'])
                paragraphs = soup.find_all('p')

                # Combine headings and relevant paragraphs
                content = []
                for heading in headings:
                    content.append(heading.get_text(strip=True))
                    next_siblings = heading.find_next_siblings()
                    for sibling in next_siblings:
                        if sibling.name == 'div':
                            content.append(sibling.get_text(strip=True))
                        elif sibling.name in ['h1', 'h2']:
                            break

                link_contents[link] = {
                    'title': title,
                    'content': '\n'.join(content)
                }
            else:
                link_contents[link] = {
                    'title': 'Failed to retrieve',
                    'content': f'Status code {response.status_code}'
                }
        except requests.RequestException as e:
            link_contents[link] = {
                'title': 'Error',
                'content': str(e)
            }

    return link_contents

# Example usage:
# links = [
#     "https://www.example.com",
#     "https://www.wikipedia.org",
#     # Add more links as needed
# ]

link_contents = get_heading_and_text_from_links(links)

# Print the extracted heading and text for each link
for link, content in link_contents.items():
    print(f"Title from {link}:\n{content['title']}\n")
    print(f"Content from {link}:\n{content['content']}\n{'-'*80}\n")
```

```python
from selenium import webdriver

def get_page_content_with_selenium(url):
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    driver = webdriver.Chrome(options=options)

    driver.get(url)
    content = driver.page_source
    driver.quit()
    return content

def get_heading_and_text_from_links(links):
    link_contents = {}

    for link in links:
        try:
            page_content = get_page_content_with_selenium(link)
            soup = BeautifulSoup(page_content, 'html.parser')

            title = soup.title.string if soup.title else 'No title found'

            headings = soup.find_all(['h1', 'h2', 'h3'])
            paragraphs = soup.find_all('p')

            content = []
            for heading in headings:
                content.append(heading.get_text(strip=True))
                next_siblings = heading.find_next_siblings()
                for sibling in next_siblings:
                    if sibling.name == 'p':
                        content.append(sibling.get_text(strip=True))
                    elif sibling.name in ['h1', 'h2', 'h3']:
                        break

            link_contents[link] = {
                'title': title,
                'content': '\n'.join(content)
            }
        except Exception as e:
            link_contents[link] = {
                'title': 'Error',
                'content': str(e)
            }

    return link_contents

# Example usage:
# links = [
#     "https://www.example.com",
#     "https://www.wikipedia.org",
#     # Add more links as needed
# ]

link_contents = get_heading_and_text_from_links(links)

for link, content in link_contents.items():
    print(f"Title from {link}:\n{content['title']}\n")
    print(f"Content from {link}:\n{content['content']}\n{'-'*80}\n")
```
